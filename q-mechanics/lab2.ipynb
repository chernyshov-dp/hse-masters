## Matrix method for solving the Schr√∂dinger equation (lab 2)

```python
import numpy as np
import matplotlib.pyplot as plt


# Harmonic oscillator
def v_harmonic(n):
    return n**2
    #return 0

# Example of the existence of a local level
def v_local(n):
    return -1 if n == 50 else 0

# Example of formation of levels in a potential well
def v_well(n):
    return -4.0 if abs(n - 500) <= 20 else 0

def create_h(size, potential_function):
    h = np.zeros((size, size))

    for n in range(size):
        h[n, n] = 2 + potential_function(n + 1)

    for n in range(size - 1):
        h[n, n + 1] = -1
        h[n + 1, n] = -1

    return h

size = 100

# Harmonic oscillator
h_harmonic = create_h(size, v_harmonic)
eig_harmonic, eve_harmonic = np.linalg.eigh(h_harmonic)

# Local level
h_local = create_h(size, v_local)
eig_local, eve_local = np.linalg.eigh(h_local)

# Potential well
size = 500
h_well = create_h(size, v_well)
eig_well, eve_well = np.linalg.eigh(h_well)

plt.figure(figsize=(14, 6))

plt.plot(eig_harmonic, marker="o", linestyle="None")
plt.xlabel("n")
plt.ylabel("En")
plt.title("Harmonic Oscillator")
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(np.sort(eig_local), marker="o", linestyle="None")
plt.xlabel("n")
plt.ylabel("En")
plt.title("Local Level")
plt.show()

plt.figure(figsize=(14, 6))
plt.plot(np.sort(eig_well), marker="o", linestyle="None")
plt.xlabel("n")
plt.ylabel("En")
plt.title("Potential Well")
plt.show()

#plt.tight_layout()
#plt.show()
```
